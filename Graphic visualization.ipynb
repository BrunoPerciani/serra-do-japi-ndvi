```python
import matplotlib.pyplot as plt

# Dados
anos = list(range(2013, 2024))
ndvi_1 = [0.6368, 0.625, 0.6501, 0.6395, 0.6543, 0.6601, 0.6724, 0.6159, 0.6165, 0.638, 0.644]
precipitacao_1 = [1342.67, 1181.88, 1655.67, 1414.02, 1397.91, 1132.26, 1376.17, 1097.22, 1039.23, 1231.17, 1273.30]

ndvi_2 = [0.8344, 0.8158, 0.8181, 0.8322, 0.8349, 0.8116, 0.8367, 0.8093, 0.8071, 0.824, 0.842]
precipitacao_2 = [1330.50, 1235.09, 1760.13, 1433.64, 1441.03, 1183.17, 1452.93, 1101.65, 1067.83, 1228.62, 1395.83]

# Plotagem
fig, ax1 = plt.subplots(figsize=(10, 6))

# NDVI
ax1.plot(anos, ndvi_1, marker='o', linestyle='-', label='Average NDVI in the municipality of Jundiaí', color='green')
ax1.plot(anos, ndvi_2, marker='^', linestyle='--', label='Average NDVI in Serra do Japi', color='darkgreen')
ax1.set_xlabel('Years of analysis')
ax1.set_ylabel('Average NDVI Index', color='black')
ax1.tick_params('y', colors='black')

# Precipitação
ax2 = ax1.twinx()
ax2.plot(anos, precipitacao_1, marker='s', linestyle='-.', markersize=5, label='Total annual precipitation in the municipality of Jundiaí', color='blue')
ax2.plot(anos, precipitacao_2,  marker='d', linestyle=':', markersize=5, label='Total annual precipitation in Serra do Japi', color='darkblue')
ax2.set_ylabel('Total annual precipitation (mm)', color='black')
ax2.tick_params('y', colors='black')

# Adicionando grade padrão
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')

# Posicionamento da legenda
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper center', bbox_to_anchor=(0.5, -0.15), shadow=True, ncol=2)

fig.tight_layout()
plt.title('Average NDVI values ​​and annual precipitation in the municipality of Jundiaí and Serra do Japi (2013 - 2023)')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados
anos = list(range(2013, 2024))
ndvi_1 = [0.6368, 0.625, 0.6501, 0.6395, 0.6543, 0.6601, 0.6724, 0.6159, 0.6165, 0.638, 0.644]
precipitacao_1 = [1342.67, 1181.88, 1655.67, 1414.02, 1397.91, 1132.26, 1376.17, 1097.22, 1039.23, 1231.17, 1273.30]

ndvi_2 = [0.8344, 0.8158, 0.8181, 0.8322, 0.8349, 0.8116, 0.8367, 0.8093, 0.8071, 0.824, 0.842]
precipitacao_2 = [1330.50, 1235.09, 1760.13, 1433.64, 1441.03, 1183.17, 1452.93, 1101.65, 1067.83, 1228.62, 1395.83]

# Plotagem
fig, ax1 = plt.subplots(figsize=(10, 6))

# NDVI
ax1.plot(anos, ndvi_1, marker='o', linestyle='-', label='NDVI médio no município de Jundiaí', color='green')
ax1.plot(anos, ndvi_2, marker='^', linestyle='--', label='NDVI médio na Serra do Japi', color='darkgreen')
ax1.set_xlabel('Anos de análise')
ax1.set_ylabel('Índice de NDVI médio', color='black')
ax1.tick_params('y', colors='black')

# Precipitação
ax2 = ax1.twinx()
ax2.plot(anos, precipitacao_1, marker='s', linestyle='-.', label='Precipitação anual total no município de Jundiaí', color='blue')
ax2.plot(anos, precipitacao_2, marker='d', linestyle=':', label='Precipitação anual total na Serra do Japi', color='darkblue')
ax2.set_ylabel('Precipitação anual total (mm)', color='black')
ax2.tick_params('y', colors='black')

# Adicionando grade padrão
plt.grid(True, linestyle='--', linewidth=0.5, color='gray')

# Posicionamento da legenda
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper center', bbox_to_anchor=(0.5, -0.15), shadow=True, ncol=2)

fig.tight_layout()
plt.title('Valores médios de NDVI e precipitação anual no município de Jundiaí e na Serra do Japi (2013 - 2023)')

plt.show()
```
